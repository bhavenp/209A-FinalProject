```python
import warnings
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#readin loan stats
loanstats_18_1 = pd.read_csv("LoanStats_2018Q1.csv",skiprows=1, encoding='latin-1')
loanstats_18_2 = pd.read_csv("LoanStats_2018Q2.csv",skiprows=1, encoding='latin-1')
loanstats_18_3 = pd.read_csv("LoanStats_2018Q3.csv",skiprows=1, encoding='latin-1')
```

```python
## descriptives
display(loanstats_18_1.describe());
display(loanstats_18_2.describe());
display(loanstats_18_3.describe());
```

```python
## checking shapes
loanstats_18_1.shape, loanstats_18_2.shape, loanstats_18_3.shape
```

```python
## checking if the columns are named the same
list(loanstats_18_1.columns) == list(loanstats_18_2.columns) == list(loanstats_18_3.columns)
```

```python
## concatenating
loanstats18 = pd.concat([loanstats_18_1, loanstats_18_2, loanstats_18_3])
```

```python
plt.hist(loanstats18.loan_amnt)
plt.title('Distribution of loan amount - 2018 (Q1 to Q3)')
plt.show()
```
