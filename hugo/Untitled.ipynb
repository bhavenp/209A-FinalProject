```python
import pandas as pd
import numpy as np

df_train = pd.read_csv("df_train", low_memory = False);
df_test = pd.read_csv("df_test", low_memory = False);
```

```python
train_loans_bought = pd.read_csv("loans_bought_2014", low_memory = False);
test_loans_bought = pd.read_csv("loans_bought_2015", low_memory = False);
```

```python
test_loan_subset = test_loans_bought.groupby('ZIP').agg({
    'ZIP' : 'count',
    'pct_accepted' : np.mean
})
```

```python
np.sum(test_loan_subset.ZIP * test_loan_subset.pct_accepted)/np.sum(test_loans_bought.shape[0])
```
